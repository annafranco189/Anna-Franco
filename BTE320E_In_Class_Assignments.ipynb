```python
sales = float(input("Enter Sales Value ($):"))
tax_rate = float(input("Enter tax rate (%):"))

cogs = 0.4 * sales
gross_profit = sales - cogs
taxes = gross_profit * tax_rate / 100
net = gross_profit - taxes


print(f"Sales: {sales}$")
print(f"Cost of Goods Sold: {cogs}$")
print(f"Gross Profit: {gross_profit}$")
print(f"Taxes: {taxes}$")
print(f"Net Income: {net}$")

```

```python
X = float(input("Enter Numerical Value:"))
Y = float(input("Enter Numerical Value:"))
Operator = input("Enter Mathemaatical Operator (+, -, *, /,//,**):")
if Operator == "+":
  print(X+Y)
elif Operator == "-":
  print(X-Y)
elif Operator == "*":
  print(X*Y)
elif Operator == "/":
  print(X/Y)
elif Operator == "//":
  print(X//Y)
elif Operator == "**":
  print(X**Y)
else:
  print("Invalid operator entered.")
```

```python
import random

choice = input("Do you want to play? (yes/no): ").lower()

while choice == "yes":
  player_move = input("Enter your move (rock, paper, scissors): ").lower()
  computer_move = random.choice(["rock","paper","scissors"])

  if player_move == computer_move:
      print("draw")
  elif (
      (player_move == "rock" and computer_move == "scissors") or
      (player_move == "paper" and computer_move == "rock") or
      (player_move == "scissors" and computer_move == "paper")
  ):
    print("you win!")
  elif (
      (player_move == "rock" and computer_move == "paper") or
      (player_move == "paper" and computer_move == "scissors") or
      (player_move == "scissors" and computer_move == "rock")):
    print("you loose")
  else:
    print("invalid move")

  choice = input("Do you want to play again? (yes/no): ").lower()

if choice == "no":
  print("Thanks for playing")
```

```python
numbers = []

choice = input("Enter a number:")

while choice != "0":
  try:
    num = float(choice)
    numbers.append(num)
  except ValueError:
    print("Invalid input")

  choice = input("Enter a number:")

if numbers:
  numbers.sort(reverse=True)
  print(numbers)
else:
  print("Invalid Input")
```

```python
numbers = []

num = int(input("Enter a number:"))

while num != 0:
  numbers.append(num)
  num = int(input("Enter a number:"))

numbers.sort(reverse=True)
print(numbers)

```

```python
name_counts = {}

while True:
    name = input("Enter a name: ")

    if name == "":
        break

    if name not in name_counts:
        name_counts[name] = 1
    else:
        name_counts[name] += 1

print(name_counts)
```
