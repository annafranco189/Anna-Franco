```python
Sales = float(input("Enter Sales Value ($):"))

COGS = 0.4 * Sales

GrossProfit = Sales - COGS

print(f"Sales: {Sales}$")
print(f"Cost of Goods Sold: {COGS}$")
print(f"Gross Profit: {GrossProfit}$")

TaxRate = float(input("Enter Tax Rate Value (%):"))
Taxes = GrossProfit * TaxRate
NetIncome = GrossProfit - Taxes

print(f"Taxes: {Taxes}$")
print(f"Net Income: {NetIncome}$")

```

```python
X = float(input("Enter Numerical Value:"))
Y = float(input("Enter Numerical Value:"))
Operator = input("Enter Mathemaatical Operator (+, -, *, /):")
if Operator == "+":
  print(X+Y)
elif Operator == "-":
  print(X-Y)
elif Operator == "*":
  print(X*Y)
elif Operator == "/":
  print(X/Y)
else:
  print("Invalid operator entered.")
```
